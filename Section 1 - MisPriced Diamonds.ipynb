### Price of Diamonds as the Size Inceases. Clarity Represents Different Qualities of the Diamonds.

```r
ggplot(data=mydata[mydata$carat<2.5,],
      aes(x=carat,
         y=price,
         color=clarity))+
geom_point(alpha=0.07)+
geom_smooth()
```
